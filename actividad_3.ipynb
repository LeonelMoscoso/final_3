# Punto_10

```python
# Paso 1: Encontrar el índice del vino con el precio más alto
indice_precio_max = datos_wine['price'].idxmax()

# Paso 2: Obtener todos los datos de ese vino usando loc
vino_mas_caro = datos_wine.loc[indice_precio_max]

# Mostrar la información (excluyendo columnas largas como 'description' si es necesario)
print(vino_mas_caro[['country', 'province', 'points', 'price', 'variety', 'winery']])
```

# Punto_11

```python
# Filtrar por país (US) y provincia (California)
California_wines = datos_wine[
(datos_wine['country'] == 'US') &
(datos_wine['province'] == 'California')
]
# Obtener las 10 variedades más frecuentes
top_variedades = California_wines['variety'].value_counts().head()

# Mostrar el resultado
print("Top 10 tipos de uva más comunes en California:")
print(top_variedades)
```
